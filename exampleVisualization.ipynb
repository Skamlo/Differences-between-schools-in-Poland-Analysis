```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import os
from dotenv import load_dotenv

load_dotenv()

host = "localhost"
database = "educational_institutions"
user = os.getenv('SQL_USER')
password = os.getenv('SQL_PASSWORD')

connection_string = f"postgresql://{user}:{password}@{host}/{database}"

%load_ext sql
%sql $connection_string
```

```python
query = 'SELECT czy_zatrudnia_psychologa, count(*) FROM rspo GROUP BY czy_zatrudnia_psychologa;'
result = %sql {query}

sns.barplot(result.DataFrame(), x="czy_zatrudnia_psychologa", y="count")
plt.title("Liczba szkół zatrudniających psychologa")
plt.xlabel("")
plt.ylabel("Liczba szkół")
plt.show()
```
